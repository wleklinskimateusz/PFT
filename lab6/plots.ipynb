```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("lab6pi4.csv")
```

```python
df
```

```python
x = df["x"]
y = df["y"]
V = df["V"]
```

```python
K = int(np.sqrt(x.shape[0]))
```

```python
K
```

```python
y.shape
```

```python
plt.imshow(np.array(V).reshape((K, K)))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Electric Potential")
plt.tight_layout()
plt.savefig("output/potentialpi4.png")
plt.show()
```

```python
Ex = np.array(df["Ex"]).reshape((K, K))
Ey = np.array(df["Ey"]).reshape((K, K))
Bx = np.array(df["Bx"]).reshape((K, K))
By = np.array(df["By"]).reshape((K, K))
```

```python
plt.quiver(x, y, Ex, Ey)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Electric field")
plt.tight_layout()
plt.savefig("output/Efieldpi4.png")
plt.show()
```

```python
plt.quiver(x, y, Bx, By)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Magnetic field")
plt.tight_layout()
plt.savefig("output/Bfieldpi4.png")
plt.show()
```
